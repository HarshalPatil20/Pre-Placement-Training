```python
import sys

class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def maxSubtreeSum(node):
    if node is None:
        return 0

    leftSum = maxSubtreeSum(node.left)
    rightSum = maxSubtreeSum(node.right)
    currentSum = node.value + leftSum + rightSum

    return max(node.value, currentSum, leftSum, rightSum)

# Example usage
root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)
root.right.left = Node(6)
root.right.right = Node(7)

maxSum = maxSubtreeSum(root)
print("Maximum subtree sum:", maxSum)
```

```python
<aside>
ðŸ’¡ Question-2

Construct the BST (Binary Search Tree) from its given level order traversal.

Example:

Input: arr[] = {7, 4, 12, 3, 6, 8, 1, 5, 10}

Output: BST:

            7

         /Â  Â  \

       4 Â  Â  12

     /Â  \ Â  Â  /

    3 Â  6Â  8

   / Â   / Â    \

 1Â    5  Â    10

</aside>
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def constructBST(levelOrder):
    if not levelOrder:
        return None

    root = Node(levelOrder[0])
    queue = [root]
    i = 1

    while i < len(levelOrder):
        current = queue.pop(0)
        leftValue = levelOrder[i]
        i += 1

        if leftValue != None:
            leftNode = Node(leftValue)
            current.left = leftNode
            queue.append(leftNode)

        if i < len(levelOrder):
            rightValue = levelOrder[i]
            i += 1

            if rightValue != None:
                rightNode = Node(rightValue)
                current.right = rightNode
                queue.append(rightNode)

    return root

# Example usage
levelOrder = [7, 4, 12, 3, 6, 8, 1, 5, 10]
bst = constructBST(levelOrder)

# Function to print the BST in inorder traversal
def inorderTraversal(node):
    if node:
        inorderTraversal(node.left)
        print(node.value, end=" ")
        inorderTraversal(node.right)

print("BST:")
inorderTraversal(bst)
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def canRepresentBST(arr):
    n = len(arr)
    if n == 0:
        return "No"  # Empty array cannot represent a BST

    stack = [Node(arr[0])]
    index = 1

    for i in range(1, n):
        node = None

        while len(stack) > 0 and arr[i] > stack[-1].value:
            node = stack.pop()

        if node:
            node.right = Node(arr[i])
            stack.append(node.right)
        else:
            node = stack[-1]
            node.left = Node(arr[i])
            stack.append(node.left)

    # Check if there are any remaining elements in the stack
    while len(stack) > 0:
        node = stack.pop()
        if node.value < arr[-1]:
            return "No"

    return "Yes"

# Example usage
arr1 = [7, 4, 12, 3, 6, 8, 1, 5, 10]
arr2 = [11, 6, 13, 5, 12, 10]

print("Output1:", canRepresentBST(arr1))
print("Output2:", canRepresentBST(arr2))
```
