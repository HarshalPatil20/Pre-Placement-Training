Given two linked list of the same size, the task is to create a new linked list using those linked lists. The condition is that the greater node among both linked list will be added to the new linked list.

Input: list1 = 5->2->3->8
list2 = 1->7->4->5
Output: New list = 5->7->4->8

Input:list1 = 2->8->9->3
list2 = 5->3->6->4
Output: New list = 5->8->9->4
</aside>

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def addNode(self, value):
        newNode = Node(value)
        if self.head is None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            self.tail = newNode

    def traverse(self):
        current = self.head
        while current:
            print(current.value, end=" ")
            current = current.next
        print()

def createNewList(list1, list2):
    newList = LinkedList()
    current1 = list1.head
    current2 = list2.head

    while current1 and current2:
        if current1.value >= current2.value:
            newList.addNode(current1.value)
        else:
            newList.addNode(current2.value)
        current1 = current1.next
        current2 = current2.next

    # Append remaining nodes from list1 or list2
    while current1:
        newList.addNode(current1.value)
        current1 = current1.next

    while current2:
        newList.addNode(current2.value)
        current2 = current2.next

    return newList

# Example usage
list1 = LinkedList()
list1.addNode(5)
list1.addNode(2)
list1.addNode(3)
list1.addNode(8)

list2 = LinkedList()
list2.addNode(1)
list2.addNode(7)
list2.addNode(4)
list2.addNode(5)

newList = createNewList(list1, list2)
newList.traverse()
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def addNode(self, value):
        newNode = Node(value)
        if self.head is None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            self.tail = newNode

    def traverse(self):
        current = self.head
        while current:
            print(current.value, end=" ")
            current = current.next
        print()

    def removeDuplicates(self):
        current = self.head
        while current and current.next:
            if current.value == current.next.value:
                current.next = current.next.next
            else:
                current = current.next

# Example usage
linkedlist = LinkedList()
linkedlist.addNode(11)
linkedlist.addNode(11)
linkedlist.addNode(11)
linkedlist.addNode(21)
linkedlist.addNode(43)
linkedlist.addNode(43)
linkedlist.addNode(60)

linkedlist.removeDuplicates()
linkedlist.traverse()
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def addNode(self, value):
        newNode = Node(value)
        if self.head is None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            self.tail = newNode

    def traverse(self):
        current = self.head
        while current:
            print(current.value, end=" ")
            current = current.next
        print()

    def reverseKNodes(self, k):
        if k == 1 or self.head is None:
            return

        def reverseLinkedList(head, tail):
            prev = None
            current = head
            while current != tail:
                nextNode = current.next
                current.next = prev
                prev = current
                current = nextNode
            return prev

        current = self.head
        prev_tail = None
        while current:
            head = current
            for _ in range(k - 1):
                if current.next:
                    current = current.next
                else:
                    break
            tail = current.next if current.next else None
            current.next = None
            reversed_head = reverseLinkedList(head, tail)
            if prev_tail:
                prev_tail.next = reversed_head
            else:
                self.head = reversed_head
            head.next = tail
            prev_tail = head
            current = tail

# Example usage
linkedlist = LinkedList()
linkedlist.addNode(1)
linkedlist.addNode(2)
linkedlist.addNode(2)
linkedlist.addNode(4)
linkedlist.addNode(5)
linkedlist.addNode(6)
linkedlist.addNode(7)
linkedlist.addNode(8)

k = 4
linkedlist.reverseKNodes(k)
linkedlist.traverse()
```

```python
<aside>
ðŸ’¡ **Question 4**

Given a linked list, write a function to reverse every alternate k nodes (where k is an input to the function) in an efficient way. Give the complexity of your algorithm.

**Example:**
Inputs:   1->2->3->4->5->6->7->8->9->NULL and k = 3
Output:   3->2->1->4->5->6->9->8->7->NULL.
</aside>
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def reverse_alternate_k_nodes(head, k):
    if not head:
        return None

    current = head
    prev = None
    next_node = None
    count = 0

    # Reverse the first k nodes
    while current and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    # Connect the reversed k nodes to the previous and next nodes
    if head:
        head.next = current

    # Skip the next k nodes
    count = 0
    while count < k - 1 and current:
        current = current.next
        count += 1

    # Recursively reverse the remaining alternate k nodes
    if current:
        current.next = reverse_alternate_k_nodes(current.next, k)

    return prev
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def addNode(self, value):
        newNode = Node(value)
        if self.head is None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            self.tail = newNode

    def traverse(self):
        current = self.head
        while current:
            print(current.value, end=" ")
            current = current.next
        print()

    def deleteLastOccurrence(self, key):
        if self.head is None:
            return

        lastOccurrenceNode = None
        prevNode = None
        current = self.head

        while current:
            if current.value == key:
                lastOccurrenceNode = current
            current = current.next

        current = self.head

        while current and current.next:
            if current.next == lastOccurrenceNode:
                prevNode = current
            current = current.next

        if lastOccurrenceNode:
            if prevNode:
                prevNode.next = lastOccurrenceNode.next
            else:
                self.head = lastOccurrenceNode.next
            lastOccurrenceNode = None

# Example usage
linkedlist = LinkedList()
linkedlist.addNode(1)
linkedlist.addNode(2)
linkedlist.addNode(3)
linkedlist.addNode(5)
linkedlist.addNode(2)
linkedlist.addNode(10)

key = 2
linkedlist.deleteLastOccurrence(key)
linkedlist.traverse()
```

```python
<aside>
ðŸ’¡ **Question 6**

Given two sorted linked lists consisting ofÂ **N**Â andÂ **M**Â nodes respectively. The task is to merge both of the lists (in place) and return the head of the merged list.

**Examples:**

Input:Â a: 5->10->15,Â b: 2->3->20

Output:Â 2->3->5->10->15->20

Input:Â a: 1->1, b: 2->4

Output:Â 1->1->2->4

</aside>
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def merge_sorted_lists(a, b):
    # If either list is empty, return the other list
    if not a:
        return b
    if not b:
        return a

    # Initialize a new head
    if a.data <= b.data:
        head = a
        a = a.next
    else:
        head = b
        b = b.next

    # Current node to track the merged list
    current = head

    # Merge the lists
    while a and b:
        if a.data <= b.data:
            current.next = a
            a = a.next
        else:
            current.next = b
            b = b.next
        current = current.next

    # Attach the remaining nodes, if any
    if a:
        current.next = a
    if b:
        current.next = b

    return head
```

```python
# Example 1
a1 = Node(5)
a2 = Node(10)
a3 = Node(15)
a1.next = a2
a2.next = a3

b1 = Node(2)
b2 = Node(3)
b3 = Node(20)
b1.next = b2
b2.next = b3

merged1 = merge_sorted_lists(a1, b1)
current = merged1
while current:
    print(current.data, end="->")
    current = current.next
# Output: 2->3->5->10->15->20

print()

# Example 2
a1 = Node(1)
a2 = Node(1)
a1.next = a2

b1 = Node(2)
b2 = Node(4)
b1.next = b2

merged2 = merge_sorted_lists(a1, b1)
current = merged2
while current:
    print(current.data, end="->")
    current = current.next
# Output: 1->1->2->4
```

```python
<aside>
ðŸ’¡ **Question 7**

Given aÂ **Doubly Linked List**, the task is to reverse the given Doubly Linked List.

**Example:**
Original Linked list 10 8 4 2
Reversed Linked list 2 4 8 10
</aside>
```

```python
class Node:
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None

class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    def addNode(self, value):
        newNode = Node(value)
        if self.head is None:
            self.head = newNode
            self.tail = newNode
        else:
            self.tail.next = newNode
            newNode.prev = self.tail
            self.tail = newNode

    def traverse(self):
        current = self.head
        while current:
            print(current.value, end=" ")
            current = current.next
        print()

    def reverseList(self):
        if self.head is None:
            return

        current = self.head
        while current:
            temp = current.prev
            current.prev = current.next
            current.next = temp
            current = current.prev

        if temp:
            self.head = temp.prev

# Example usage
linkedlist = DoublyLinkedList()
linkedlist.addNode(10)
linkedlist.addNode(8)
linkedlist.addNode(4)
linkedlist.addNode(2)

linkedlist.reverseList()
linkedlist.traverse()
```

```python
<aside>
ðŸ’¡ **Question 8**

Given a doubly linked list and a position. The task is to delete a node from given position inÂ a doubly linked list.

**Example 1:**
Input:
LinkedList = 1 <--> 3 <--> 4
x = 3
Output:1 3
Explanation:After deleting the node at
position 3 (position starts from 1),
the linked list will be now as 1->3.


</aside>
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.prev = None
        self.next = None

def delete_node(head, position):
    if not head:
        return head

    # If position is 1, update the head
    if position == 1:
        new_head = head.next
        if new_head:
            new_head.prev = None
        return new_head

    # Traverse to the node at the given position
    current = head
    count = 1
    while current and count < position:
        current = current.next
        count += 1

    # If position is out of bounds, return the original head
    if not current:
        return head

    # Update the pointers of the previous and next nodes
    prev_node = current.prev
    next_node = current.next
    if prev_node:
        prev_node.next = next_node
    if next_node:
        next_node.prev = prev_node

    return head
```

```python
# Example 1
node1 = Node(1)
node3 = Node(3)
node4 = Node(4)
node1.next = node3
node3.prev = node1
node3.next = node4
node4.prev = node3

position = 3

new_head = delete_node(node1, position)
current = new_head
while current:
    print(current.data, end=" ")
    current = current.next
# Output: 1 3
```
